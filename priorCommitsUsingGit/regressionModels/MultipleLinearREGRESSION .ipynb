```python
import pandas as pd
import math
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
data = pd.read_csv(r"C:\Users\rajpr\Downloads\archive\diabetes-dataset.csv")

X = data[['Pregnancies','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']].values
y = data['Glucose'].values

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

reg = LinearRegression().fit(X, y)
#mlp.fit(x_train, np.ravel(y_train))
reg.score(X, y)

####
mlrTrain_predictY = reg.predict(x_train)
mlrTest_predictY = reg.predict(x_test)

mlr_trainR2 = r2_score(y_train, mlrTrain_predictY )
mlr_testR2 = r2_score(y_test, mlrTest_predictY)

print("Performance of the MLR model on training model (R^2) is: ", mlr_trainR2)
print("Performance of the MLR model on testing model (R^2) is:", mlr_testR2)

mlrTrain_RMSE = math.sqrt(mean_squared_error(y_train, mlrTrain_predictY))
mlrTest_RMSE = math.sqrt(mean_squared_error(y_test, mlrTest_predictY))

print("Training Data RMSE (accuracy for regression): ", mlrTrain_RMSE)
print("Testing Data RMSE (accuracy for regression): ", mlrTest_RMSE)

print("MLR predicted y for training: ", mlrTrain_predictY)
print("MLR predicted y for testing: ", mlrTest_predictY )

#print("Mean absolute error: ", mean_absolute_error(y_test, y_pred))
#print("Mean sqared error: ", mean_squared_error(y_test, y_pred))
#print("Root mean squared error: ", np.sqrt(mean_squared_error(y_test, y_pred)))
#print("R^2 score: ", lr.score(x_test, y_test))
```
