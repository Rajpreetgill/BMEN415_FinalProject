```python
# blood pressure (predictor 1)
# patient age (predictor 2)
# skin thickness (predictor 3)
# insulin levels (predictor 4) 
# glucose level measurement (target outcome)
from sklearn.metrics import r2_score
import math
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
x = data.iloc[1:,[2,3,4,7]] #predictors
y = data.iloc[1:,1].values #outcome
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)

# Fit regression model
n_neighbors = 2

knn = neighbors.KNeighborsRegressor(n_neighbors)
knn.fit(x_train, y_train)

knnTrain_predictY = knn.predict(x_train)
knnTest_predictY = knn.predict(x_test)

knn_trainR2 = r2_score(y_train, knnTrain_predictY)
knn_testR2 = r2_score(y_test, knnTest_predictY)

print("Performance of the k nearest neighbor model on training model (R^2) is: ", knn_trainR2)
print("Performance of the k nearest neighbor model on testing model (R^2) is:", knn_testR2)

knnTrain_RMSE = math.sqrt(mean_squared_error(y_train, knnTrain_predictY))
knnTest_RMSE = math.sqrt(mean_squared_error(y_test, knnTest_predictY))

print("Training Data RMSE (accuracy for regression): ", knnTrain_RMSE)
print("Testing Data RMSE (accuracy for regression): ", knnTest_RMSE)

print("knn predicted y for training: ", knnTrain_predictY)
print("knn predicted y for testing: ", knnTest_predictY)

#plt.subplot(2, 1, i + 1)
#plt.scatter(x_test[:,0], y_test[:,0], color="blue", label="data")
#plt.plot(y, y_train, color="navy", label="prediction")
#plt.axis("tight")
#plt.legend()
#plt.title("KNeighborsRegressor (k = %i, weights = '%s')" % (n_neighbors, weights))

#plt.tight_layout()
#plt.show()
```
